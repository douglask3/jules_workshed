### Simulated Burnt Area Time Series

In this analysis, we simulate burnt area over a period of 20 years with monthly data points. The time series incorporates two key components:

1. **Seasonal Cycle:** This represents the regular, periodic fluctuations in burnt area due to seasonal factors. We model this with a sinusoidal function, creating peaks and troughs that repeat each year.
2. **Trend:** An upward trend is added to represent the gradual increase in burnt area over time, potentially simulating the effects of long-term climate changes or land-use practices.

Two versions of the burnt area time series are generated:

- **Burnt Area with Trend:** This series includes both the seasonal cycle and an increasing trend over time.
- **Burnt Area without Trend:** This version contains only the seasonal cycle, with the trend component removed. It allows us to analyze the seasonal variability independently of the long-term increase.

### Visualizations

1. **Time Series Plot:** This plot shows both time series (with and without the trend) on the same graph, allowing us to visualize how the burnt area changes over time due to seasonal and trend components. The time series with the trend is shown in blue, while the time series without the trend is shown in orange.

2. **Histogram:** The histogram illustrates the distribution of burnt area values in each time series. The distribution with the trend (in blue) shows higher burnt area values due to the added trend, while the distribution without the trend (in orange) highlights the range of values solely from seasonal variation.

These visualizations allow us to understand both the overall increase in burnt area over time and the regular, cyclical variations within each year.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate time series data
np.random.seed(42)  # For reproducibility
n_years = 20
months_per_year = 12
total_months = n_years * months_per_year
time = pd.date_range(start="2000-01-01", periods=total_months, freq="M")

# Generate seasonal cycle and trend
seasonal_cycle = 100 + 50 * np.sin(2 * np.pi * (time.month - 1) / 12)  # A sinusoidal seasonal cycle

trend = np.linspace(0, 200, total_months)  # Increasing trend over time

# Combine trend, seasonal cycle, and some noise
burnt_area = seasonal_cycle + trend + np.random.normal(scale=20, size=total_months)
seasonal_only_burnt_area = seasonal_cycle + np.random.normal(scale=20, size=total_months)

# Add the new data to the DataFrame


# Create DataFrame
df = pd.DataFrame({"Date": time, "Burnt Area": burnt_area})
df["Burnt Area (No Trend)"] = seasonal_only_burnt_area

random_months = df.sample(n=3, random_state=42)
df.loc[random_months.index, "Burnt Area"] = 380

# Plotting both time series
plt.figure(figsize=(12, 6))
plt.plot(df["Date"], df["Burnt Area"], label="Burnt Area with Trend", color="blue")
plt.plot(df["Date"], df["Burnt Area (No Trend)"], label="Burnt Area without Trend", color="orange")
plt.title("Simulated Burnt Area with and without Trend")
plt.xlabel("Year")
plt.ylabel("Burnt Area (arbitrary units)")
plt.legend()
plt.grid(True)
plt.show()
```

### Histogram of Burnt Area

In this section, we create histograms to visualize the distribution of burnt area values for both the time series with and without the trend.

The **histogram** shows how often different burnt area values occur throughout the dataset:

- **Burnt Area with Trend** (in blue): This histogram represents the distribution of burnt area values when both the seasonal cycle and upward trend are included. Higher burnt area values are more frequent due to the trend.
- **Burnt Area without Trend** (in orange): This histogram shows the distribution of values when only the seasonal cycle is present, highlighting the range of burnt area values due to seasonal variation alone.

The overlapping histograms provide a comparison of the burnt area values with and without the trend, giving insight into how the trend shifts the distribution to higher values over time.


```python
plt.figure(figsize=(10, 6))
plt.hist(df["Burnt Area"], bins=30, color="skyblue", edgecolor="black", alpha=0.7, label="With Trend")
plt.hist(df["Burnt Area (No Trend)"], bins=30, color="orange", edgecolor="black", alpha=0.5, label="Without Trend")
plt.title("Histogram of Simulated Burnt Area with and without Trend")
plt.xlabel("Burnt Area (arbitrary units)")
plt.ylabel("Frequency")
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

### Biomass vs Burnt Area with Anomalous Points

In this code, we simulate a relationship between biomass and burnt area with an overall trend of increasing burnt area as biomass increases. This setup allows us to visualize a scenario where higher biomass generally leads to more burnt area, while also introducing a few anomalously high burnt area values at lower biomass levels.

#### Steps in the Code:

1. **Simulating Biomass Data:** We generate biomass values with a normal distribution, centered around 100, representing varying biomass amounts.
2. **Generating Burnt Area Data:** Burnt area values are created with a noisy linear relationship to biomass, suggesting that areas with higher biomass tend to have more potential for burning.
3. **Adding Anomalous Points:** To represent unusual conditions, we add a few anomalously high burnt area values specifically for low biomass values. These points stand out as unusually high for their given biomass level, while still remaining lower than the highest burnt areas at higher biomass values.

#### Visualization:

- **Scatter Plot:** The plot displays the relationship between biomass and burnt area. 
  - Most points (in blue) follow a general trend where burnt area increases with biomass.
  - The anomalous points (in red) highlight cases where burnt area is high despite lower biomass, simulating unusual conditions.
  
This plot helps illustrate how anomalies can appear in a dataset with an overall trend, representing cases where fire intensity is disproportionately high at unexpectedly low biomass levels.


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic biomass data
np.random.seed(42)
biomass = np.random.normal(loc=100, scale=20, size=200)  # Centered around 100 with some variation

# Generate burnt area data that generally scales with biomass
burnt_area = 2 * biomass + np.random.normal(scale=15, size=200)  # Slightly noisy linear relationship

# Introduce a few anomalously high burnt area points at lower biomass values
anomalous_indices = np.random.choice(np.where(biomass < 75)[0], size=5, replace=False)  # Select points with low biomass
burnt_area[anomalous_indices] += 80  # Boost burnt area for these anomalies

# Plot biomass vs burnt area, highlighting anomalous points
plt.figure(figsize=(10, 6))
plt.scatter(biomass, burnt_area, color="skyblue", label="Normal Data")
plt.scatter(biomass[anomalous_indices], burnt_area[anomalous_indices], color="red", label="Anomalous High Burnt Area")
plt.title("Biomass vs Burnt Area with Anomalous Points")
plt.xlabel("Biomass (arbitrary units)")
plt.ylabel("Burnt Area (arbitrary units)")
plt.legend()
plt.grid(True)
plt.show()
```

### Histogram Analysis of Burnt Area with Anomalous Points

This section provides a histogram of burnt area values, highlighting a subset of anomalously high burnt area values that occur at low biomass levels.

#### Code Explanation

1. **Synthetic Data Generation**: Biomass and burnt area values are generated with an overall trend, where burnt area typically increases with biomass. The burnt area values include some random noise to simulate natural variability.
2. **Anomalous Data Points**: To simulate unusual fire behavior, we introduce a few anomalously high burnt area values associated with lower biomass values. These points are artificially elevated to stand out from the general trend.
3. **Histogram Visualization**: 
   - The histogram displays the distribution of burnt area values, with the general data shown in blue.
   - Anomalous points are highlighted in red to indicate burnt areas that are unexpectedly high for the given biomass level.

This visualization allows us to examine the overall distribution of burnt area while easily identifying anomalies. The highlighted points show how unusual conditions may cause high fire occurrences even when biomass is relatively low.


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic biomass and burnt area data
np.random.seed(42)
biomass = np.random.normal(loc=100, scale=20, size=200)  # Biomass centered around 100
burnt_area = 2 * biomass + np.random.normal(scale=15, size=200)  # Burnt area with noise

# Introduce a few anomalously high burnt area points at lower biomass values
anomalous_indices = np.random.choice(np.where(biomass < 90)[0], size=5, replace=False)  # Select points with low biomass
burnt_area[anomalous_indices] += 50  # Boost burnt area for these anomalies

# Plot histogram of burnt area with anomalous points highlighted in red
plt.figure(figsize=(10, 6))
# Plot histogram for normal data
plt.hist(burnt_area, bins=20, color="skyblue", edgecolor="black", alpha=0.7, label="Normal Burnt Area")

# Overlay histogram for anomalous points in red
plt.hist(burnt_area[anomalous_indices], bins=20, color="red", edgecolor="black", alpha=0.7, label="Anomalous High Burnt Area")

plt.title("Histogram of Burnt Area with Anomalous Points Highlighted")
plt.xlabel("Burnt Area (arbitrary units)")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()
```

# Detecting Anomalous Burnt Areas Using Isolation Forest

This notebook demonstrates how to use the Isolation Forest algorithm to identify anomalies in burnt area data, with temperature and precipitation as contextual variables. By focusing on burnt area while considering temperature and precipitation effects, we can detect unusual fire events that deviate from expected environmental patterns.

---

## Steps

### 1. Generate Synthetic Data
We create synthetic data for precipitation, temperature, and burnt area, introducing anomalies in burnt area values to test the model's ability to detect them.

### 2. Visualize the Data
A scatter plot of precipitation vs. burnt area shows some points marked as anomalies. This provides an overview of burnt area variation with precipitation, highlighting the introduced anomalies.

### 3. Isolation Forest Model Setup
- We initialize the Isolation Forest algorithm with `contamination=0.05`, indicating we expect around 5% of data points to be anomalous.
- The model is trained on three variables: `Precipitation`, `Temperature`, and `Burnt_Area`, allowing it to learn the relationship between burnt area and environmental conditions.

### 4. Anomaly Detection
- The model is used to predict anomalies, labeling each data point as either normal (`1`) or anomalous (`-1`). 
- It detects burnt area values that deviate significantly from expected patterns based on temperature and precipitation.

### 5. 3D Visualization of Anomalies
A 3D scatter plot shows:
- **Blue points** representing normal data.
- **Red points** highlighting detected anomalies where burnt area values deviate from the expected relationship with precipitation and temperature.

This plot provides insight into burnt area patterns under various environmental conditions, helping identify unusual fire events.

---

## Code
The code below implements these steps:

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate synthetic data for demonstration
np.random.seed(42)
n_samples = 1000

# Precipitation (e.g., in mm)
precip = np.random.normal(loc=100, scale=20, size=n_samples)

# Temperature (e.g., in °C)
temp = np.random.normal(loc=25, scale=5, size=n_samples)

# Burnt area (e.g., in hectares), with normal values and a few anomalies
burnt_area = np.random.normal(loc=50, scale=10, size=n_samples)

# Introduce anomalies in burnt area for certain conditions
burnt_area[::50] = burnt_area[::50] + np.random.normal(50, 10, n_samples // 50)

# Combine into a DataFrame
data = pd.DataFrame({'Precipitation': precip, 'Temperature': temp, 'Burnt_Area': burnt_area})

# Visualize the synthetic data
plt.figure(figsize=(10, 5))
plt.scatter(data['Precipitation'], data['Burnt_Area'], alpha=0.5, label="Normal Data")
plt.scatter(data['Precipitation'][::50], data['Burnt_Area'][::50], color='red', label="Anomalies")
plt.xlabel("Precipitation (mm)")
plt.ylabel("Burnt Area (ha)")
plt.legend()
plt.title("Synthetic Burnt Area Data with Anomalies")
plt.show()

# Prepare data for Isolation Forest
# We use Burnt Area as the main variable and also include Precipitation and Temperature as contextual information
features = data[['Precipitation', 'Temperature', 'Burnt_Area']].values

# Initialize Isolation Forest with contamination set to 5% (i.e., we expect around 5% anomalies)
isolation_forest = IsolationForest(contamination=0.05, random_state=42)
isolation_forest.fit(features)

# Predict anomalies
# Anomalies are labeled -1, normal points are labeled 1
data['Anomaly'] = isolation_forest.predict(features)

# Separate normal and anomalous data for visualization
anomalies = data[data['Anomaly'] == -1]
normal_data = data[data['Anomaly'] == 1]

# Plot the 3D scatter plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Normal data points (blue)
ax.scatter(normal_data['Precipitation'], normal_data['Temperature'], normal_data['Burnt_Area'],
           c='blue', label='Normal Data', alpha=0.5)

# Anomalous data points (red)
ax.scatter(anomalies['Precipitation'], anomalies['Temperature'], anomalies['Burnt_Area'],
           c='red', label='Detected Anomalies', alpha=0.8)

# Labels and title
ax.set_xlabel('Precipitation (mm)')
ax.set_ylabel('Temperature (°C)')
ax.set_zlabel('Burnt Area (ha)')
ax.set_title("3D Plot of Burnt Area Anomalies Detected by Isolation Forest")
ax.legend()

plt.show()
```
